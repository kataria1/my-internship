Split the data

```python
x=df.drop('vote',axis=1)
y=df['vote']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x=st.fit_transform(x)
x
```

```python
x=pd.DataFrame(x)
x
```

```python
lr=LinearRegression()
```

```python
for i in range(0,60):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    
    print('at random state',i)
    print('the training accuracy is',r2_score(y_train,pred_train))
    print('the testing accuracy is',r2_score(y_test,pred_test))
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=57)
```

```python
lr.fit(x_train,y_train)
pred_train=lr.predict(x_train)
pred_test=lr.predict(x_test)
print(r2_score(y_train,pred_train))
print(r2_score(y_test,pred_test))
```

at this there is not very much difference b/w training and testing model 

```python
from sklearn.model_selection import GridSearchCV
from  sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso()
values={'alpha':[0.1,0.001,0.0001]}
```

```python
gr=GridSearchCV(ls,values)
gr.fit(x,y)
print(gr)
```

```python
print(gr.best_params_)
```

```python
ls=Lasso(alpha=0.1)
ls.fit(x_train,y_train)
pred_test=ls.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
rd=Ridge()
values={'alpha':[0.1,0.001,0.0001,1]}
```

```python
gr=GridSearchCV(rd,values)
gr.fit(x,y)
print(gr)
```

```python
print(gr.best_params_)
```

```python
rd=Ridge(alpha=1)
rd.fit(x_train,y_train)
pred_test=rd.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred_train=dtr.predict(x_train)
pred_test=dtr.predict(x_test)
print(r2_score(y_train,pred_train))
print(r2_score(y_test,pred_test))
```

```python
from sklearn.ensemble import RandomForestRegressor 
rf=RandomForestRegressor(n_estimators=200,random_state=53)
rf.fit(x_train,y_train)
pred_train=rf.predict(x_train)
pred_test=rf.predict(x_test)
print(r2_score(y_test,pred_test))
```

as the random forest we are getting high r2_score so it is our final model
