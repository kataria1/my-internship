separating the columns into input and output 

```python
x=df.drop('TotalCharges',axis=1)
y=df['TotalCharges']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x=st.fit_transform(x)
x
```

```python
x=pd.DataFrame(data=x)
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression()
```

```python
for i in range(0,60):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    
    print('at random state',i)
    print('the training accuracy is',r2_score(y_train,pred_train))
    print('the testing accuracy is',r2_score(y_test,pred_test))
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=43)
```

```python
lr.fit(x_train,y_train)
pred_test=lr.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for j in range(2,10):
    cv=cross_val_score(lr,x,y,cv=j)
    cv=cv.mean()
    print('at random state',j)
    print('cross_val_score',cv)
    print('\n')
```

at cv=6 best result getting

Regularization

```python
from sklearn.model_selection import GridSearchCV
from  sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso()
values={'alpha':[0.1,0.001,0.0001]}
```

```python
gr=GridSearchCV(ls,values)
gr.fit(x,y)
print(gr)
```

```python
print(gr.best_params_)
```

```python
ls=Lasso(alpha=0.1,random_state=43)
ls.fit(x_train,y_train)
pred_test=ls.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
cv=cross_val_score(ls,x,y,cv=6)
cv=cv.mean()
print('r2_score',r2_score(y_test,pred_test))
print('cross_score',cv)
```

```python
rd=Ridge()
values={'alpha':[0.1,0.001,0.0001,1]}
```

```python
gr=GridSearchCV(rd,values)
gr.fit(x,y)
print(gr)
```

```python
print(gr.best_params_)
```

```python
rd=Ridge(alpha=1,random_state=43)
rd.fit(x_train,y_train)
pred_test=rd.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
cv=cross_val_score(rd,x,y,cv=6)
cv=cv.mean()
print('r2_score',r2_score(y_test,pred_test))
print('cross_score',cv)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
k={'kernel':['rbf','linear','poly']}
```

```python
gr=GridSearchCV(svr,k)
gr.fit(x,y)
print(gr)
```

```python
print(gr.best_params_)
```

```python
svr=SVR(kernel='linear')
svr.fit(x_train,y_train)
pred_test=svr.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
cv=cross_val_score(svr,x,y,cv=6)
cv=cv.mean()
print('r2_score',r2_score(y_test,pred_test))
print('cross_score',cv)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf=RandomForestRegressor(n_estimators=200,random_state=43)
rf.fit(x_train,y_train)
pred_test=rf.predict(x_test)
print(r2_score(y_test,pred_test))
```

```python
cv=cross_val_score(rf,x,y,cv=5)
cv=cv.mean()
print('r2_score',r2_score(y_test,pred_test))
print('cross_score',cv)
```

we are getting r2_score and cross validation both as 99.8% which shows our model is performing well.
