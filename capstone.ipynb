```python
driver=webdriver.Chrome(r'C:\Users\ADMIN\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url="https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
driver.get(url)
time.sleep(2)
```

```python
serial=[]
s= driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]') 
for i in s:
    serial.append(i.text)
#print(serial)
print(len(serial))
```

```python
movies=[]
m= driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a")
for i in m:
    movies.append(i.text)
#print(movies)
print(len(movies))
```

```python
director_names = []
director_elements = driver.find_elements(By.XPATH,"//div[@class='lister-item mode-advanced']//p[@class='']")
for element in director_elements:
    director_name = element.find_element(By.TAG_NAME,'a')
    director_names.append(director_name.text)
print(len(director_names))
#print(director_names)
```

```python
duration=[]
d= driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in d:
    duration.append(i.text)
#print(duration)
print(len(duration))
```

```python
year=[]
y=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in y:
    year.append(i.text)    
#print(year)
print(len(year))
```

```python
rating=[]
r= driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
for i in r:
    rating.append(i.text)
#print(rating)
print(len(rating))
```

```python
metascore=[]
m= driver.find_elements(By.XPATH,'//span[@class="metascore  favorable"]')
for i in m:
    metascore.append(i.text)
print(len(metascore))
```

```python
import csv
import pandas as pd
from itertools import zip_longest
```

```python
csv1 = zip_longest(serial, movies, director_names, duration, year, rating,metascore, fillvalue="N/A")

filename = 'csv1.csv'


with open(filename, 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['serial', 'movies', 'director_names', 'duration', 'year', 'rating','metascore'])  
    writer.writerows(csv1)

```

```python
movies=[]
m= driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a")
for i in m:
    movies.append(i.text)
#print(movies)
print(len(movies))
```

```python
vote = []
v = driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]')
for element in v:
    vote_text = element.text.split('|')[0].replace('Votes:', '').strip()
    vote_count = int(vote_text.replace(',', ''))
    vote.append(vote_count)

#print(vote)
print(len(vote))
```

```python
stars_list= []

r= driver.find_elements(By.CSS_SELECTOR, '.lister-item-content')
for result in r:
    stars = result.find_elements(By.CSS_SELECTOR, '.lister-item-content .ghost~ a')
    star = [star.text for star in stars]
    star = ', '.join(star)
    stars_list.append(star)
#print(stars_list)
print(len(stars_list))
```

```python
genre_list = []
genre_element = driver.find_elements(By.CSS_SELECTOR, '.lister-item-content .genre')
for i in genre_element:
    genre_list.append(i.text.strip())

#print(genre_list)
print(len(genre_list))
```

```python
gross_list = []
gross_elements = driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]/span[contains(text(),"Gross:")]')
for element in gross_elements:
    gross = element.find_element(By.XPATH, 'following-sibling::span[1]').text.strip()
    gross = gross.replace('$', '').replace('M', '')  
    gross_value = float(gross)  
    gross_list.append(gross_value)

#print(gross_list)
print(len(gross_list))
```

```python
popularity=[]
p= driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
for i in p:
    popularity.append(i.text)
#print(popularity)
print(len(popularity))
```

```python
certificate=[]
c= driver.find_elements(By.XPATH,'//span[@class="certificate"]')
for i in c:
    certificate.append(i.text)
#print(certificate)
print(len(certificate))
```

```python
csv2 = zip_longest(movies, stars_list, vote, genre_list, gross_list, popularity, certificate, fillvalue="N/A")

filename = 'csv2.csv'


with open(filename, 'w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['movies', 'stars_list', 'vote', 'genre_list', 'gross_list', 'popularity','certificate'])  
    writer.writerows(csv2)
```
